```python
import pandas as pd
df_microdados = pd.read_csv('microdados_ed_basica_2022.csv', delimiter=';', encoding='ISO-8859-1')
df_microdados = df_microdados[df_microdados['SG_UF'] == 'SP']
df_microdados = df_microdados[(df_microdados['NO_MUNICIPIO'] == 'São Paulo')]
df_microdados.head()
```

```python
df_rotatividade = pd.read_excel('rotatividade.xlsx', sheet_name='Planilha2')
df_rotatividade.head()
print(len(df_rotatividade))
```

```python
# Selecionando as colunas relevantes do df_rotatividade
df_rotatividade = df_rotatividade[['Nome Escola', 'Pontuação IRD']]

# Renomeando a coluna 'NO_ENTIDADE' para 'Nome Escola' no df_microdados
df_microdados.rename(columns={'NO_ENTIDADE': 'Nome Escola'}, inplace=True)

# Mesclando os DataFrames com base na coluna 'Nome Escola'
df_merged = pd.merge(df_microdados, df_rotatividade, on='Nome Escola', how='left')

# Criando um novo DataFrame com apenas as colunas de interesse
df_novo = df_merged[['Nome Escola', 'Pontuação IRD']].dropna().sort_values(by='Pontuação IRD', ascending=True)

# Exibindo as primeiras linhas do novo DataFrame
df_novo.head()
```
